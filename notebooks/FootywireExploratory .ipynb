```python
shared_venues = {'MCG': ['Collingwood', 'Hawthorn', 'Melbourne', 'Richmond'], 
                 'Adelaide Oval': ['Port Adelaide', 'Adelaide'],
                 ' Marvel Stadium': ['Essendon', 'Carlton', 'North Melbourne', 'Western Bulldogs', 'St Kilda'],
                 'Optus Stadium': ['Fremantle', 'West Coast']}

def home_away(x):
    if x[2] in shared_venues:
        if x[0] in shared_venues[f'{x[2]}'] and x[1] in shared_venues[f'{x[2]}']:
            return False
        else:
            return True
    else:
        return True
games_adv = games.copy()
games_adv['Home_Away'] = games_adv.apply(lambda row: [row['Home.Team'], row['Away.Team'], row['Venue']], axis=1)
games_adv['HomeVAway'] = games_adv['Home_Away'].apply(home_away)
games_adv = games_adv.drop(columns='Home_Away')

true_home_away = games_adv[games_adv['HomeVAway'] == True]

scores = true_home_away[['Home.Points', 'Away.Points']]

scores_melted = scores.melt(var_name='TeamType', value_name='Score')

plt.figure(figsize=(8,6))
sns.boxplot(data=scores_melted, x='TeamType', y='Score')
plt.title("Distribution of Home vs Away Team Scores")
plt.xlabel("")
plt.ylabel("Score")
plt.show()
```

No change

```python

home_scores = games_adv[games_adv['HomeVAway'] == True][['Home.Team', 'Home.Points']].rename(columns={'Home.Team': 'Team', 'Home.Points': 'Score'})
home_scores['GameType'] = 'Home'

away_scores = games_adv[games_adv['HomeVAway'] == True][['Away.Team', 'Away.Points']].rename(columns={'Away.Team': 'Team', 'Away.Points': 'Score'})
away_scores['GameType'] = 'Away'

team_scores = pd.concat([home_scores, away_scores], ignore_index=True)

team_scores_diff = team_scores.groupby(by=['Team','GameType'], as_index=False)['Score'].mean()
team_scores_diff.loc[team_scores_diff['GameType'] == 'Away', 'Score'] *= -1
team_scores_diff.groupby('Team', as_index=False)['Score'].sum().sort_values(by='Score', ascending=False).reset_index(drop=True).rename(columns={'Score': 'Ave Home Extra Points'})
```
